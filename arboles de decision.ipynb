```python
feature_names = list(map(str, x.columns.values))
class_names = list(map(str, y.unique()))

fig = plt.figure(figsize=(16, 9))
tree.plot_tree(arbol, feature_names=feature_names, class_names=class_names, filled=True)
plt.show()
```
